## Defining our Binary Search Function

```python
def locate_cards(cards, query):
    lo = 0
    hi = len(cards) - 1
    flag = False
    while lo <= hi:
        mid = (lo + hi) // 2
        
        print("lo:", lo, " hi:",hi, " mid:", mid)
        
        #numbers are arranged in decreasing order
        
        if cards[mid] == query:
            flag = True
            return mid
        elif cards[mid] > query:
            lo = mid + 1
        elif cards[mid] < query:
            hi = mid - 1
    
    
    # or only return -1
    if flag == False:
        return -1
```

```python
def test_location(cards, query, mid):
    mid_number = cards[mid]
    
    if mid_number == query:
        if mid - 1 >= 0 and cards[mid - 1] == query:
            return 'left'
        else: 
            return 'found'
        
    elif mid_number < query:
        return 'left'
    else:
        return 'right'

def locate_cards(cards, query):
    lo, hi = 0, len(cards) - 1
    
    while lo <= hi:
        print("lo:", lo, " hi:",hi)
        
        mid = (lo + hi) // 2
        
        result = test_location(cards, query, mid)
        
        #numbers are arranged in decreasing order
        
        if result == 'found':
            return mid
        elif result == 'right':
            lo = mid + 1
        elif result == 'left':
            hi = mid - 1
    

    
    return -1    
```
